```python
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np

# Load Raw Audio Data into [sig] and [sr]
sig, sr = sf.read('Lightroom2.raw', channels=2, samplerate=62500, format='RAW', subtype='PCM_16')

# Define an empty array to store the combined audio channels
data = []

# Average the left and right audio channels and store in [data]
for i in range(0, len(sig)):
    average = (sig[i][0]+sig[i][1])/2
    data.append(average)

# Create an array to store the tick location on the x axis
start = 0
end = len(data)
label_location = np.arange(start, end, 300)

# Set the plot size
plt.figure(figsize=(100,20))
# Put vertical ticks at every item in [label_location]
plt.xticks(label_location, rotation='vertical')
# Plot the data
plt.plot(data)
# Save the data to an image file
## plt.savefig("Adobe1.svg", format="svg")
plt.savefig('Lightroom2.png', dpi=100)
# Show the plot
plt.show()
```
